```python
# Load Processed data
import pandas as pd

df = pd.read_csv("../data/processed/netflix_processed.csv")

df.head()
```

```python
# ðŸ¥ˆ Step 4 â€” Load Transformer Model (DL Model)
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

# This is a pretrained transformer neural network.

# This line alone = Deep Learning.
```

```python
# ðŸ¥‰ Step 5 â€” Generate Embeddings
embeddings = model.encode(df["combined_features"], show_progress_bar=True)


# This converts every movie description into a neural vector representation.

# It may take 1â€“3 minutes depending on system.
```

```python
# ðŸ§  Step 6 â€” Cosine Similarity on Embeddings
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim_dl = cosine_similarity(embeddings)
```

```python
# ðŸŽ¯ Step 7 â€” Deep Learning Recommendation Function
def recommend_movies_dl(title):
    if title not in df["title"].values:
        return ["Movie not found"]

    idx = df[df["title"] == title].index[0]
    sim_scores = list(enumerate(cosine_sim_dl[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]

    movie_indices = [i[0] for i in sim_scores]

    return df["title"].iloc[movie_indices].reset_index(drop=True)
```

```python
# ðŸ§ª Test It
recommend_movies_dl("Inception")


# Now you have:

# âœ” ML version
# âœ” DL version

# You can compare outputs.
```

```python
# ðŸ† STEP 8 â€” Save DL Model

# Add:

import pickle

pickle.dump(embeddings, open("../models/dl_embeddings.pkl", "wb"))
pickle.dump(df, open("../models/movies_dl.pkl", "wb"))

print("Deep Learning model saved successfully!")


# Now your models/ folder will have DL files too.
```
