```python
#Step 1 â€” Load Cleaned Data

import pandas as pd
import re

df = pd.read_csv("../data/raw/netflix_movies_detailed_up_to_2025.csv")

# Keep only required columns
df = df[["title", "genres", "description", "vote_average", "popularity"]]

# Drop missing & duplicates
df = df.dropna()
df = df.drop_duplicates()

df.shape
```

```python
'''
ðŸ¥ˆ Step 2 â€” Clean Text Columns

We clean:

Lowercase

Remove special characters

Remove extra spaces
'''
def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()

df["genres"] = df["genres"].apply(clean_text)
df["description"] = df["description"].apply(clean_text)
```

```python
# Create Combined Features
df["combined_features"] = df["genres"] + " " + df["description"]

df[["title", "combined_features"]].head() 

# Now every movie has one unified text representation.
```

```python
# Step 4 â€” Save Processed Dataset - it create separately in the proceesed folder - cleandata
df.to_csv("../data/processed/netflix_processed.csv", index=False)
# Now your clean dataset is saved separately.
```
