```python
# Load Processed Data
import pandas as pd

df = pd.read_csv("../data/processed/netflix_processed.csv")

df.head()
```

```python
# ðŸ¥ˆ Step 2 â€” TF-IDF Vectorization

# This converts text â†’ numbers.

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(df["combined_features"])

tfidf_matrix.shape
```

```python
# ðŸ¥‰ Step 3 â€” Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

cosine_sim.shape

#That means every movie compared with every movie.
```

```python
#ðŸ§  Step 4 â€” Recommendation Function
def recommend_movies(title, cosine_sim=cosine_sim):
    idx = df[df["title"] == title].index[0]

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:6]  # Top 5 similar

    movie_indices = [i[0] for i in sim_scores]

    return df["title"].iloc[movie_indices]
```

```python
recommend_movies("Inception")

# If it gives similar movies â†’ YOU BUILT A RECOMMENDATION ENGINE ðŸ”¥
```

```python
import pickle

# Save vectorizer
pickle.dump(tfidf, open("../models/tfidf.pkl", "wb"))

# Save cosine similarity
pickle.dump(cosine_sim, open("../models/cosine_sim.pkl", "wb"))

# Save cleaned dataframe
pickle.dump(df, open("../models/movies.pkl", "wb"))

print("Models saved successfully!")
```
